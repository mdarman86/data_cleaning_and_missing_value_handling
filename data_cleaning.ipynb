## Dropping Columns with High Missing Values
Columns having more than 40% missing values are removed.

```python
threshold = 0.4
missing_ratio = df.isnull().mean()

cols_to_drop = missing_ratio[missing_ratio > threshold].index
df.drop(columns=cols_to_drop, inplace=True)

print("Dropped columns:", list(cols_to_drop))
```

## Missing Values Visualization
This bar Chart shows the number of missing values in each column

```python
import matplotlib.pyplot as  plt

missing = df.isnull().sum()
missing = missing[missing > 0]

missing.plot(kind='bar')
plt.title("Missing Values per Column")
plt.xlabel("Columns")
plt.ylabel("Count")
plt.show()
```


Some column Contain a High number of missing values,
While others have very few or none.

Visualize missing data patterns using simple bar Charts.

Numerical columns were filled with the mean or median, and categorical
columns were filled with the most common value.
